```python
# 결측값 처리 예시 = 평균 대체(수치형 데이터) = 데이터가 비어있는 부분을 해당 열의 전체 평균값으로 채워 넣는다
# 이상치(Outlier)에 영향을 많이 받는다(평균이 왜곡될 수 있다)

df['Sales'] = df['Sales'].fillna(df['Sales'].mean())
df.isnull().sum()
```

```python
# 결측값 처리 예시 = 최빈값(범주형 변수) = 결측값을 해당 열에서 가장 빈번하게 나타나는 값으로 채우는 기법
# 범주형 변수(예: 색상, 성별, 도시 등)는 평균/중앙값처럼 숫자로 대표값을 구하기 어렵다
# 그래서 가장 많이 나온 값(최빈값, mode)으로 결측치를 채우면, 데이터 분포를 크게 왜곡하지 않고 결측값을 합리적으로 대체할 수 있다
# 이상치(Outlier)에 영향을 받지 않는다
# mode() = 최빈값, 즉 컬럼에서 가장 많이 나온 값을 반환 = 반환 형태가 Series이므로 최빈값이 여러개라도 [0]으로 첫 번째 값을 선택한다


df['Category'] = df['Category'].fillna(df['Category'].mode()[0])
df.isnull().sum()
```

```python
# Discount 결측치 처리(결측치가 너무 많으면 평균이나 최빈값을 넣는 것 보다 열 자체를 삭제하는게 좋을 수 있다)

df = df.drop(columns = ['Discount'])
df.head(3)
```

```python
# 이상치 - IQR (Quantity 단일 변수) 이상치 확인

import numpy as np

q1 = np.percentile(df['Quantity'], 25)
q2 = np.percentile(df['Quantity'], 50)
q3 = np.percentile(df['Quantity'], 75)
iqr = q3 - q1

lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)

outliers = df[(df['Quantity'] < lower_bound) | (df['Quantity'] > upper_bound)]

print("Lower Bound: ", lower_bound)
print("Upper Bound: ", upper_bound)
print()
print(outliers.head(5))
print()
print("Q1: ", q1)
print("Q2: ", q2)
print("Q3: ", q3)
print()
print("IQR: ", iqr)
```

```python
plt.figure(figsize = (7, 5))
plt.boxplot(df['Quantity'], vert = True)

plt.axhline(y = q1, color = '#50C878', linestyle = '--', label = f'Q1 : {q1:.2f}')
plt.axhline(y = q3, color = '#40E0D0', linestyle = '--', label = f'Q3 : {q3:.2f}')
plt.axhline(y = lower_bound, color = '#008080', linestyle = ':', label = f'Lower Bound : {lower_bound:.2f}')
plt.axhline(y = upper_bound, color = '#800080', linestyle = ':', label = f'Upper Bound : {upper_bound:.2f}')

plt.title("IQR(Quantity)")
plt.ylabel("Quantity")
plt.legend()
plt.show()
```

정규화/표준화

정규화

변수들끼리 크기 차이가 있을 때 서로 영향을 최소한으로 줄이기 위해 값의 범위를 0 - 1로 조정하는 것

머신러닝 전처리나 시각화 시 값의 크기를 비교하기 쉽게 만든다

변수 크기 차이가 너무 클 때 사용 이상치가 너무 큰게 있을 때도 사용

표준화

평 0 → 평균(평균값)이 0이라는 뜻 → 데이터를 평균이 0이 되도록 이동시킴

표 1 → 표준편차(standard deviation)가 1이라는 뜻 → 데이터를 표준편차가 1이 되도록 스케일링

즉, 표준화를 거친 데이터는 평균이 0이고, 표준편차가 1인 분포가 됩니다.

```python
# 정규화 (Normalization) - MinMax

df['sales_norm'] = (df['sales'] - df['sales'].min()) / (df['sales'].max() - df['sales'].min())
df[['sales', 'sales_norm']].head(2)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df['sales_norm'] = scaler.fit_transform(df[['sales']])
```

```python
# 표준화

df['profit_std'] = (df['profit'] - df['profit'].mean()) / df['profit'].std()
df[['profit', 'profit_std']].head(2)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df['profit_std'] = scaler.fit_transform(df[['profit']])
```

```python
# 머지는 키 값을 줘야된다

left = pd.DataFrame({
    "id": [1, 2, 3],
    "score": [90, 85, 88]
})

right = pd.DataFrame({
    "id": [2, 3, 4],
    "grade": ["B", "A", "C"]
})
```

```python
left
```

```python
right
```

```python
#pd.merge() → 두 데이터프레임을 열 기준으로 합치는 함수
# left, right → 합칠 두 데이터프레임, 기준이 되는걸 먼저 써준다
# on='id' → 공통 열(id)을 기준으로 합침 = 공통컬럼, 합치는 것에 대한 기준
# how='inner' → **교집합 방식(inner join)**으로 합침
# left (왼쪽 기준) 왼쪽 데이터프레임 기준으로 모두 남김 오른쪽에 없는 값은 NaN으로 채움 두 데이터프레임 모두에 있는 id만 결과에 남음
# right (오른쪽 기준) 오른쪽 데이터프레임 기준으로 모두 남김 왼쪽에 없는 값은 NaN으로 채움
# outer (합집합) 양쪽 데이터프레임의 모든 값 포함 어느 한쪽에 없는 값은 NaN

merged = pd.merge(left, right, on = 'id', how = 'inner')
```

```python
merged_l = pd.merge(left, right, on = 'id', how = 'left')
merged_l
```

```python
merged_r = pd.merge(left, right, on = 'id', how = 'right')
merged_r
```

```python
merged_o = pd.merge(left, right, on = 'id', how = 'outer')
merged_o
```

```python
# join(): 인덱스 기준 병합

# 인덱스 설정

left_indexed = left.set_index('id')
right_indexed = right.set_index('id')
```
