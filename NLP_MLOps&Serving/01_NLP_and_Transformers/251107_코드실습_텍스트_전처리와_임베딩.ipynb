```python
from gensim.models import Word2Vec
from konlpy.tag import Okt

sentence = [
    "나는 밥을 먹었다",
    "오늘 날씨가 좋다",
    "밥 먹고 산책하러 갔다",
    "고양이는 귀엽다",
    "강아지는 충성스럽다"
]

okt = Okt()
tokenized = [okt.morphs(s) for s in sentence]

model = Word2Vec(sentences = tokenized, vector_size = 50, window = 3, min_count = 1, sg = 1)
print("'밥' 벡터: ", model.wv['밥'])
print(model.wv.most_similar("밥", topn = 3))
print("'고양이' + '귀엽다'와 유사한 단어: ")
print(model.wv.most_similar(positive = ['고양이', '귀엽다'], topn = 3))
```
