```python
# 막대 그래프 - 데이터가 뚝뚝 떨어진 데이터들 = 범주형 데이터(X)를 비교할 때 좋다

# y = count

categories = ["A", "B", "C", "D"]
values = np.array([23, 45, 12, 31])

plt.bar(categories, values, color = '#1C6BA0', edgecolor = 'black', linewidth = 1.7, width = 0.7)
plt.title('막대 그래프')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
#산점도 - 두 값의 '관계'를 시각화할 때 사용

x = np.random.rand(100)
y = x + np.random.normal(0, 0.1, 100)

plt.scatter(x, y, alpha = 0.6, color = 'teal', s = 40)
plt.title('산점도')
plt.xlabel('X')
plt.ylabel('Y', rotation = 45)
plt.xticks(rotation = 45)
plt.show()
```

```python
# 히스토그램

data = np.random.normal(0, 1, 1000)

plt.hist(data, bins = 30, color = 'turquoise', density = True)
# density = True: 확률밀도 그래프로 변경
plt.title('히스토그램')
plt.show()
```

```python
# 파이 차트

y = ["A", "B", "C", "D"]
sizes = [25, 30, 20, 25]

# 수치를 넣고 labels로 이름표를 붙여준다
plt.pie(sizes, labels = y)
plt.title("파이 차트")
plt.show()
```

```python
y = ["A", "B", "C", "D"]
sizes = [25, 30, 20, 25]
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
# 파이차트에서 어느 부분 하나만 강조할 때
explode = (0, 0.1, 0, 0)

"""
수치가 가장 먼저 들어가야된다,
몰아 넣을건 몰아 넣는다,
autopct: 파이조각 위에 퍼센트를 보여준다,
startangle: 시작 각도 조절,
shadow = True: 그림자 효과
"""

plt.pie(
    sizes,
    labels = y,
    colors = colors,
    autopct = "%1.1f%%",
    startangle = 90,
    explode = explode,
    shadow = True,
    counterclock = False,
    wedgeprops={'edgecolor': 'black', 'linewidth': 1}
)
plt.title('커스터 마이징된 파이차트', fontsize = 17, fontweight = 'bold')
# 그래프 겹쳐보이지 않게 해준다
plt.tight_layout()
plt.show()
```

```python
# 다중 레이아웃 - subplot(행, 열, 그래프 순번) 기본

x = np.linspace(0, 2 * np.pi, 100)

plt.figure(figsize = (6, 3))

plt.subplot(1, 2, 1)
plt.plot(x, np.sin(x))
plt.title("사인 그래프")

plt.subplot(1, 2, 2)
plt.plot(x, np.cos(x))
plt.title("사인 그래프")

plt.tight_layout()
plt.show()
```

```python
# 다중 레이아웃 - subplot(행, 열, 그래프 순번) 응용
# subplot은 0부터 시작하지 않는다
# plt.legend()는 그래프에 범례(legend)를 추가하는 함수

x = np.linspace(0, 2 * np.pi, 100)

plt.figure(figsize = (7, 7))

for i in range(9):
    plt.subplot(3, 3, i + 1)
    plt.plot(x, np.sin(x + i), label = f"sin(x + {i})")
    plt.title(f"{i + 1} 사인 그래프")
    plt.xticks([])
    plt.yticks([])
    plt.legend()

plt.tight_layout()
plt.show()
```
